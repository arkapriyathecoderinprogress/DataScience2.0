## Filtering a DataFrame

```python
# find all the ipl winners
mask=ipl['MatchNumber']=='Final'
new_df=ipl[mask]
new_df[['Season','WinningTeam']]
```

```python
ipl[ipl['MatchNumber']=='Final'][['Season','WinningTeam']]
```

```python
# how many super over finishes have occured
ipl[ipl['SuperOver']=='Y'].shape[0]
```

```python
# how many matches has CSK won in Kolkata
ipl[(ipl['City']=='Kolkata') & (ipl['WinningTeam']=='Chennai Super Kings')].shape[0] # bitwise OR operator 2 boolean series
```

```python
# toss winner is match winner in percentage
(ipl[ipl['TossWinner']==ipl['WinningTeam']].shape[0]/ipl.shape[0])*100
```

```python
# movies with rate higher than 8 and votes>10000
movies[(movies['imdb_rating']>8) & (movies['imdb_votes']>10000)]
```

```python
movies[(movies['imdb_rating']>8) & (movies['imdb_votes']>10000)].shape[0]
```

```python
# action movies with rating higher than 7.5
m1=movies['genres'].str.split('|').apply(lambda x:'Action' in x)
m2=movies['imdb_rating']>7.5
movies[m1 & m2]
```

```python
m1=movies['genres'].str.contains('Action')
m2=movies['imdb_rating']>7.5
movies[m1 & m2]
```
